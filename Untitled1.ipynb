```python
X=train.drop(['id','hour_bef_precipitation', 'hour_bef_visibility', 'hour_bef_pm10','hour_bef_pm2.5','hour_bef_ozone','count'], axis=1)
y =train['count']

X1=test.drop(['id','hour_bef_precipitation', 'hour_bef_visibility', 'hour_bef_pm10','hour_bef_pm2.5','hour_bef_ozone'], axis=1)
```

```python
from sklearn.linear_model import LinearRegression #    Linear Regression  .

model = LinearRegression() #  .
model.fit(X, y) #  .
```

```python
y_hat = model.predict(X) # y .
```

```python
nmae = np.mean(abs(y_hat - y) / y) #  nmae .
print(f' NMAE: {nmae}')
```

```python
plt.style.use('ggplot') # ggplot   .
plt.figure(figsize=(20, 10)) #     .
plt.plot(y_hat, label = 'prediction') #     .
plt.plot(y, label = 'real') #     .
plt.legend(fontsize = 20) #    .
plt.show() #   .
```

```python
from sklearn.metrics import mean_squared_error
print("RMSE : ",mean_squared_error(y_hat,y)**0.5)
```
