## **linear regression (선형 회귀)**

종속 변수 y와 한 개 이상의 독립 변수 (또는 설명 변수) X와의 선형 상관 관계를 모델링하는 회귀분석 기법

**파일 불러오기**

```python
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import files
uploaded=files.upload()
```

```python
train=pd.read_csv('train.csv') 
test=pd.read_csv('test.csv') 
submission = pd.read_csv('submission.csv')
```

**train, test 데이터 결측값 처리**

```python
train['hour_bef_temperature'] = train['hour_bef_temperature'].fillna(value = train['hour_bef_temperature'].mean())
train['hour_bef_ozone'] = train['hour_bef_ozone'].fillna(value = train['hour_bef_ozone'].mean())
train['hour_bef_windspeed'] = train['hour_bef_windspeed'].fillna(value = train['hour_bef_windspeed'].mean())
train['hour_bef_humidity'] = train['hour_bef_humidity'].fillna(value = train['hour_bef_humidity'].mean())
```

```python
test['hour_bef_temperature'] = test['hour_bef_temperature'].fillna(value = test['hour_bef_temperature'].mean())
test['hour_bef_ozone'] = test['hour_bef_ozone'].fillna(value = test['hour_bef_ozone'].mean())
test['hour_bef_windspeed'] = test['hour_bef_windspeed'].fillna(value = test['hour_bef_windspeed'].mean())
test['hour_bef_humidity'] = test['hour_bef_humidity'].fillna(value = test['hour_bef_humidity'].mean())
```

**데이터 전처리**

```python
X=train.drop(['id','hour_bef_precipitation', 'hour_bef_visibility', 'hour_bef_pm10','hour_bef_pm2.5','count'], axis=1)
y =train['count']

X1=test.drop(['id','hour_bef_precipitation', 'hour_bef_visibility', 'hour_bef_pm10','hour_bef_pm2.5'], axis=1)
```

**모델 정의 및 학습**

```python
from sklearn.linear_model import LinearRegression # 머신러닝 중 지도학습 Linear Regression 패키지를 불러옵니다.

model = LinearRegression() # 모델를 정의합니다.
model.fit(X, y) # 학습을 시킵니다.
```

```python
y_hat = model.predict(X) # y값을 예측합니다.
y_hat
```

```python
from sklearn.metrics import mean_squared_error
print("RMSE : ",mean_squared_error(y_hat,y)**0.5)
```

**test 데이터 예측 > submission 추가**

```python
y1_hat = model.predict(X1) 
submission['count']=y1_hat
submission['count']
```

```python
plt.style.use('ggplot') # ggplot 시각화 패키지를 불러옵니다.
plt.figure(figsize=(20, 10)) # 그림의 가로 세로를 조정하는 코드입니다.
plt.plot(y_hat, label = 'prediction') # 예측 항목의 라벨을 붙이는 코드입니다.
plt.plot(y, label = 'real') # 실제 항목의 라벨을 붙이는 코드입니다.
plt.legend(fontsize = 20) # 글씨 크기를 조정하는 코드입니다.
plt.show() # 그림을 보여주는 코드입니다.
```
