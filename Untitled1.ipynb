```python
test['hour_bef_temperature'] = test['hour_bef_temperature'].fillna(value = test['hour_bef_temperature'].mean())
test['hour_bef_ozone'] = test['hour_bef_ozone'].fillna(value = test['hour_bef_ozone'].mean())
test['hour_bef_windspeed'] = test['hour_bef_windspeed'].fillna(value = test['hour_bef_windspeed'].mean())
test['hour_bef_humidity'] = test['hour_bef_humidity'].fillna(value = test['hour_bef_humidity'].mean())
```

```python
X=train.drop(['id','hour_bef_precipitation', 'hour_bef_visibility', 'hour_bef_pm10','hour_bef_pm2.5','hour_bef_ozone','count'], axis=1)
y =train['count']

X1=test.drop(['id','hour_bef_precipitation', 'hour_bef_visibility', 'hour_bef_pm10','hour_bef_pm2.5','hour_bef_ozone'], axis=1)
```

```python
from sklearn.linear_model import LinearRegression # 머신러닝 중 지도학습 Linear Regression 패키지를 불러옵니다.

model = LinearRegression() # 모델를 정의합니다.
model.fit(X, y) # 학습을 시킵니다.
```

```python
y_hat = model.predict(X) # y값을 예측합니다.
```

```python
nmae = np.mean(abs(y_hat - y) / y) # 모델 nmae를 계산합니다.
print(f'모델 NMAE: {nmae}')
```

```python
plt.style.use('ggplot') # ggplot 시각화 패키지를 불러옵니다.
plt.figure(figsize=(20, 10)) # 그림의 가로 세로를 조정하는 코드입니다.
plt.plot(y_hat, label = 'prediction') # 예측 항목의 라벨을 붙이는 코드입니다.
plt.plot(y, label = 'real') # 실제 항목의 라벨을 붙이는 코드입니다.
plt.legend(fontsize = 20) # 글씨 크기를 조정하는 코드입니다.
plt.show() # 그림을 보여주는 코드입니다.
```

```python
from sklearn.metrics import mean_squared_error
print("RMSE : ",mean_squared_error(y_hat,y)**0.5)
```
